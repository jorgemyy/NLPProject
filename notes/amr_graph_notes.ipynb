This is a jupyter notebook to get amrlib working and understand how to use it. It is not structured and more used as scratch work.

```python
import amrlib
stog = amrlib.load_stog_model()
```

```python
graphs = stog.parse_sents(['He flies to Hawaii'])
graph = graphs[0]
print(graph)
print(type(graph))
```

```python
print(type(graphs))
print(type(graphs[0]))
print(graphs[0])
```

```python
## use manipulatable penman graph
import penman
```

```python
penman.decode(graphs[1])
```

```python
g = penman.decode(graph)
print(type(g))
g.epidata
```

```python
print(sorted(g.variables()))
```

```python
for e in g.edges():
    print(e.source)
    print(e.role)
    print(e.target)
```

```python
print(g.top)
```

```python
for x,y,z in g.instances():
    print(x)
    print(z)
```
