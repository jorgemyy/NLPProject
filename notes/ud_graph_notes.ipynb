This is a jupyter notebook to get stanza working and understand how to use it. It is not structured and more used as scratch work.

```python
import stanza
stanza.download('en') # download English model
nlp = stanza.Pipeline('en') # initialize English neural pipeline
doc = nlp("Barack Obama was born in Hawaii.") # run annotation over a sentence
```

```python
# doc object
print(f"Raw Sentence: {doc.text}")
print(type(doc))
```

```python
# sentence object
print(doc.sentences[0])
```

```python
doc.sentences[0].print_dependencies
```

```python
doc.sentences[0].constituency.children
```

```python
doc.sentences[0].tokens[3].pretty_print
```

```python
# can add properties
doc.add_property('character_count',default=0,getter=lambda self:len(self.text),setter=None)
```
